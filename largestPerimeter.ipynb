```python
import itertools
def largestPerimeter(A: 'List[int]') -> 'int':
        b = list(itertools.combinations(A, 3))
        list1 = [0]
        for i in b:
            c = sorted(i)
            if c[0] + c[1] > c[2]:
                list1.append(sum(c))
                
            
        return max(list1)


print(largestPerimeter([88125,286539,464267,285127,462300,788506,798011,803544,337627,623018,995974,677530,905583,
                        562961,660060,334324,133237,629356,561284,812560,563508,269257,225282,716972,960016,817214,
                        804450,379804,336223,469324,630225,385094,998035,711808,675410,861346,46655,367296,682001,
                        96863,11139,443563,615868,887723,201034,816851,378744,715339,353191,66387,461912,847035,
                        992048,389213,949976,466666,786091,191131,180655,333248,980536,83767,218497,710536,564151,
                        168434,155833,951112,142502,139959,639711,812790,273515,538058,267338,510677,882273,136773,
                        837683,789373,394296,726775,869754,427958,918430,573984,860827,870834,451915,768278,779042,
                        483572,214107,237344,399447,860598,581868,315321,904529,490173,74716,619791,583586,537392,
                        470684,902067,640224,403109,771123,292528,402418,970134,541939,827141,622303,401658,546306,
                        817577,935935,414967,518834,965066,238679,630034,716312,656289,411207,96261,468933,563910,
                        41266,195048,698061,934475,304736,621697,947535,82270,156745,937260,854420,156941,780745,
                        552135,550929,695708,275655,840457,403100,408526,794509,691879,108673,689607,203353,729902,
                        584469,17411,636735,696067,70643,911690,936440,747259,195994,615764,532820,885752,41970,69162,
                        149724,976485,792913,477715,487619,823164,175848,933477,393166,740853,142169,785441,606838,
                        235241,761442,47545,488245,180778,694933,980834,945889,129626,820884,822323,638979,973487,
                        962714,112513,350265,338933,218977,564453,215417,659019,978435,757926,543840,529223,91718,
                        353734,799260,271287,252309,22888,614573,329959,459106,473623,818415,588400,394303,636116,
                        361234,858398,607751,860742,828722,205177,210780,269295,957583,503257,628802,275710,802064,
                        400902,966531,466513,438824,757161,411435,127577,260748,730365,803776,905100,810813,541552,
                        973894,687090,518867,474074,574112,117236,282001,81264,539360,873361,891249,741928,280713,
                        134670,361493,172696,737895,803376,181345,56549,65681,339477,268972,378564,808836,632235,
                        801185,619322,483549,72550,224810,260849,950008,875877,622005,708353,117761,233196,263725,
                        513437,58303,659006,834047,391743,200460,833696,976949,447508,732541,278130,801021,394721]))
```
